## Stationarity vs Non-stationarity

There are 2 major types of TS : stationary and non-stationary

A time series is said to be “stationary” if it has no trend, exhibits constant variance over time, and has a constant autocorrelation structure over time (its properties do not depend on the time at which the serie is observed)

```python
# We can plot the moving average or moving variance and see if it varies with time
# here we can see that the std is mostly constant

rolmean = data['Close'].rolling(5).mean()
rolstd = data['Close'].rolling(5).std()

plt.plot(data['Close'], label='hist_price')
plt.plot(rolmean, c='orange', label='moving mean 1w')
plt.plot(rolstd, c='blue', label='moving std 1w')
plt.title("Meta stock price")
plt.xticks(rotation=45, ha='right')
plt.legend();
```

### ADF Test

One way to test whether a time series is stationary is to perform an augmented Dickey-Fuller test:
- H0: The time series is non-stationary. In other words, it has some time-dependent structure and does not have constant variance over time.
- Ha: The time series is stationary

```python
adfuller(data['Close'])
```

Since the p-value is not less than 0.05, we fail to reject the null hypothesis. This means the TS is non-stationary : it has some time-dependent structure and does not have constant variance over time

```python
def test_adfuller(data, seuil):
    test_result = adfuller(data)
    print(f"ADF test statistic: {test_result[0]:.2f}")
    print(f"p-value: {test_result[1]:.2f}")
    if test_result[1] < seuil:
        print("we can reject the null hypothesis: the distribution is stationary\n")
    else:
        print("we cannot reject the null hypothesis: the distribution is not stationary\n")
```

### KPSS test

A key difference from ADF test is that the null hypothesis of the KPSS test is that the series is stationary. If p-value is < signif level (say 0.05), then the series is non-stationary.

```python
kpss(data['Close'])
```

p-value <0.05 so the time serie is non-stationnary

```python
# In KPSS test, to turn ON the stationarity testing around a trend, you need to explicitly pass the regression='ct'
kpss(data['Close'], regression='ct')
```

A major difference between KPSS and ADF tests is the capability of the KPSS test to check for stationarity in the ‘presence of a deterministic trend’ : the test may not necessarily reject the null hypothesis (that the series is stationary) even if a series is steadily increasing or decreasing.

The word ‘deterministic’ implies the slope of the trend in the series does not change permanently : even if the series goes through a shock, it tends to regain its original path.

Both tests conclude that the serie is not stationary = the serie is not stationary.

```python
def test_kpss(data, seuil, regression='c'):
    test_result = kpss(data, regression)
    print(f"KPSS test statistic: {test_result[0]:.2f}")
    print(f"p-value: {test_result[1]:.2f}")
    if regression == 'c':
        if test_result[1] > seuil:
            print("we cannot reject the null hypothesis: the distribution is stationary around a constant\n")
        else:
            print("we can reject the null hypothesis: the distribution is not stationary around a constant\n")
    else:
        if test_result[1] > seuil:
            print("we cannot reject the null hypothesis: the distribution is stationary around a trend\n")
        else:
            print("we can reject the null hypothesis: the distribution is not stationary around a trend\n")
```

## Convert non-stat time series into stat time series

Algorithms can't handle non-stationarity : they need static relationships as they build relationships between inputs and outputs by estimating the core parameters of the underlying distributions.

Non-stationary time series have unique properties such as seasonality and trend, which cause the mean and variance of the series to fluctuate, making it hard to model their behavior. Making a distribution stationary is a strict requirement in time series forecasting (extra preprocessing steps)

There are 2 major reasons behind non-stationarity : trend (varying mean over time) & seasonality (variations # specific time-frames)

The idea is to model the trend/seasonality of the series and remove them from the series to get a stationary series.

### Detrending

It involves removing the trend effects from the given dataset and showing only the differences in values from the trend. It always allows cyclical patterns to be identified.

The simplest way to detrend a time series is by subtracting the mean value of the data. 
This is called a constant model, and it assumes that the trend of the time series is a straight horizontal line.

```python
detrended = detrend(data['Close'], type='constant')
detrended = pd.Series(detrended, index=data.index)
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(data['Close'])
ax[0].set_title('Normal')
ax[0].set_xticks(ax[0].get_xticks(), ax[0].get_xticklabels(), rotation=45, ha='right')
ax[1].plot(detrended)
ax[1].set_title('Detrended')
ax[1].set_xticks(ax[1].get_xticks(), ax[1].get_xticklabels(), rotation=45, ha='right');
```

```python
test_adfuller(detrended, 0.05)
test_kpss(detrended, 0.05)
```

### We could look at the moving average to estimate/model trend (and then remove it from the series)

#### Simple Moving Average

SMA calculates the unweighted mean of the previous M or N points. We prefer selecting sliding window data points based on the amount of smoothing, as increasing the value of M or N improves smoothing but reduces accuracy.

```python
data['SMA_20'] = data.Close.rolling(20, min_periods=1).mean()
data['SMA_60'] = data.Close.rolling(60, min_periods=1).mean()
data['SMA_120'] = data.Close.rolling(120, min_periods=1).mean()

plt.plot(data['Close'], c='blue', label='historic_price')
plt.plot(data['SMA_20'], c='orange', label='1m_ma')
plt.plot(data['SMA_60'], c='green', label='3m_ma')
plt.plot(data['SMA_120'], c='gold', label='6m_ma')
plt.xticks(rotation=45, ha='right')
plt.legend();
```

```python
# let's substract the 1m MVA
data['Close-20'] = data['Close']-data['SMA_20']
plt.plot(data['Close'], c='blue', label='historic_price')
plt.plot(data['SMA_20'], c='orange', label='1m_ma')
plt.plot(data['Close-20'], c='black', label='standard')
plt.legend()
plt.xticks(rotation=45, ha='right');
```

```python
test_adfuller(data['Close-20'], 0.05)
```

## ACF and PACF

Two useful tools in time series analysis to understand temporal dependencies between data.
- ACF (autocorrelation function) : measures the correlation btw a sequence and itself at different time period ; it assesses the similarity between observations as a function of the time lag between them
- PACF (partial autocorrelation function) : correlation that excludes the effcet of intermediate terms ; it takes into account values at shorter intervals
-- ex : for a lag of k, PACF is the correlation between the ts and istelf lagged by k periods, after subtracting the effects of lags 1 to k-1

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8))

plot_acf(data['Close'], lags=30, zero=True, ax=ax1)
ax1.set_title('ACF')
ax1.set_xlabel('Lag')
ax1.set_ylabel('Correlation')
ax1.set_xticks(np.arange(0, 31, 1))

plot_pacf(data['Close'], lags=30, zero=True, ax=ax2)
ax2.set_title('PACF')
ax2.set_xlabel('Lag')
ax2.set_ylabel('Partial correlation')
ax2.set_xticks(np.arange(0, 31, 1))

plt.tight_layout();
```

Slow decrease of ACF indicates that the time series is not stationary (as previously indicated by the statistical tests)

## ARIMA Model

The ARIMA model combines autoregression (AR), moving average (MA) and differentiation (I) to account for the trend in the data (not need to convert non stationary into stationary model with ARIMA):
- AR : linear regression of current values on previous values in the time series ; capture the effects of linear dependence on past values
- MA : uses moving average of previous errors to model the relationships between residuals and current observations in the time series ; captures dependencies effects between residuals
- I : differentiation used to make the time series stationary bu removing trends and seasonal behaviors (diff between consecutive obs until a stationary series is obtained)

- p(AR) : how many past values are used to predict the current value
- d (diff) : how many times the series must be differentiated to make the data stationary
- q (MA) : how many past residuals are used to predict the current observation

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8))

plot_acf(data['Close'].diff().dropna(), lags=20, zero=True, ax=ax1)
ax1.set_title('ACF with differnciated data')
ax1.set_xlabel('Lag')
ax1.set_ylabel('Correlation')
ax1.set_xticks(np.arange(0, 21, 1))

plot_pacf(data['Close'].diff().dropna(), lags=20, zero=True, ax=ax2)
ax2.set_title('PACF with differentiated data')
ax2.set_xlabel('Lag')
ax2.set_ylabel('Partial Correlation')
ax2.set_xticks(np.arange(0, 21, 1))

plt.tight_layout();
```

```python
# pick ARIMA parameters based on observations
p = 2 # cf. PACF (1st out of window)
d = 1 # see diff above : with one diff, we got to a stationary ts
q = 2 # cf. ACF (1st out of window)
```

### Backtesting to improve confidence

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test['Target'], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    # start value = certain amount of data to train first model (2500 days = 10 years)
    # step value = nb of days to predict based on model first model (11th year)

    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
horizons = [2,5,60,250,1000]
new_pred = []
for horizon in horizons:
    rolling_av = sp500.rolling(horizon).mean()

    ratio_col = f'Close_ratio_{horizon}'
    sp500[ratio_col] = sp500['Close']/rolling_av['Close']

    trend_col = f'Trend_{horizon}'
    sp500[trend_col] = sp500.shift(1).rolling(horizon).sum()['Target']

    new_pred += [ratio_col, trend_col]
```

```python
# how many do i want to look back on to predict
# more sensitive to fluctuation if decrease
prediction_days = 60 
```

```python
x_train = []
y_train = []

for x in range(prediction_days, len(scaled_data)):
    x_train.append(scaled_data[x-prediction_days:x,0])
    y_train.append(scaled_data[x,0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1],1))
```

## Neural Prophet

```python
def predict_next_x_days(ticker, nbdays=60):
    
    df = yf.download(tickers=ticker, period='5y')
    df = df[['Close']]
    df.reset_index(inplace=True)
    df.columns = ['ds', 'y'] # must have two columns: 'ds' which has the timestamps and 'y' column which contains the observed values of the time series

    model = NeuralProphet()
    model.fit(df)

    future = model.make_future_dataframe(df, periods=nbdays)
    forecast = model.predict(future)
    actual_data = model.predict(df)

    return df, forecast, actual_data, model
```

```python
def outliers_IQR(df):
    q3, q1 = q3, q1 = np.percentile(df, [75,25])
    iqr = q3 - q1
    outliers = df[((df < (q1 - 1.5 * iqr))|(df > (q3 + 1.5 * iqr)))]
    return outliers
```

```python
def df_outliers(df):
    df_outlier = pd.DataFrame(columns = ['Date', 'Title', 'Link', 'polarity'])
    outliers = outliers_IQR(df['polarity'])
    i=0
    for title in outliers.index:
        df_outlier.loc[i]= [df.iloc[title]['Date'], df.iloc[title]['Title'], df.iloc[title]['Link'], df.iloc[title]['polarity']]
        i+=1
    df_outlier.sort_values(by='polarity', ascending=False)

    return (df_outlier)
```

```python
def boxplot_outliers(df):
    fig, axs = plt.subplots(figsize=(3,4))
    axs = df[['polarity']].boxplot()
    outliers = outliers_IQR(df['polarity'])
    for title in outliers.index:
        axs.text(x=1.02,y=outliers[title]-0.02, s=df.iloc[title]['Title'], fontsize=11, color="blue")
```

## Wordcloud

```python
import spacy
from wordcloud import WordCloud
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def lemmatize(text):
    
    text = text.lower() # tout en minuscule
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not (token.is_stop or token.is_punct or token.is_digit or len(token)<=2)]
    
    return ' '.join(tokens)
```

```python
def wordcloud_analysis(ticker):
    
    stock = finvizfinance(ticker, verbose=0)
    news = stock.ticker_news()
    news["processed_title"] = news["Title"].apply(lambda txt:lemmatize(txt))
    
    tick = yf.Ticker(ticker)
    to_delete = lemmatize(tick.info['shortName'])
    list_to_delete = to_delete.split(' ')
    list_to_delete.append("stock")
    list_to_delete.append(tick.info['symbol'].lower())

    corpus = ' '.join(news["processed_title"].tolist())
    word_list = corpus.split(' ')
    word_list_updated = [w for w in word_list if w not in list_to_delete]

    wordcloud = WordCloud(random_state=42,normalize_plurals=False, width=600, height=300, max_words=300, background_color="white", colormap='Set2')
    wordcloud.generate(" ".join(word_list_updated))
    fig, ax = plt.subplots(1,1, figsize=(9,6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"Word cloud des news relatives à {tick.info['shortName']} \n ({(news['Date'].max()-news['Date'].min()).days} derniers jours)", fontsize=16)
    plt.axis("off");
```
